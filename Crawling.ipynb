```python
import requests
import pandas as pd
from io import StringIO

URLS = {
    "학생식당": "https://www.kumoh.ac.kr/ko/restaurant01.do",
    "교직원식당": "https://www.kumoh.ac.kr/ko/restaurant02.do",
    "분식당": "https://www.kumoh.ac.kr/ko/restaurant04.do",
}

def fetch_html(url: str) -> str:
    res = requests.get(url, timeout=15)
    res.raise_for_status()
    res.encoding = "utf-8"
    return res.text

print("대상 식당:", ", ".join(URLS.keys()))
```

```python
# 각 식당 URL에서 식단표를 읽어 DataFrame으로 저장
menus = {}

for name, url in URLS.items():
    html = fetch_html(url)
    tables = pd.read_html(StringIO(html))  # FutureWarning 회피

    if not tables:
        print(f"[{name}] 테이블 없음")
        continue

    df = tables[0].copy()
    df.columns = [str(c).strip() for c in df.columns]

    # 셀 텍스트의 줄바꿈/중복 공백 정리
    df = df.replace(r"\s+", " ", regex=True)
    menus[name] = df

    print(f"[{name}] 테이블 {len(tables)}개 발견 / shape={df.shape}")

# 출력 생략 방지 옵션
pd.set_option("display.max_colwidth", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)

# 개별 결과 확인
student_df = menus.get("학생식당")
staff_df = menus.get("교직원식당")
snack_df = menus.get("분식당")

print("\n=== 학생식당 (전체) ===")
display(student_df)

print("\n=== 교직원식당 (전체) ===")
display(staff_df)

print("\n=== 분식당 (전체) ===")
display(snack_df)
```
